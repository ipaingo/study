## subplot

Графики в *matplotlib* находятся внутри объекта *Figure*. 
Для того, чтобы получить доступ к объекту, можно воспользоваться методом `plt.figure()`.

Внутри объекта *Figure* можно создавать подграфики с использованием метода `add_subplot()`.



````python
fig = plt.figure()  # Инициализируем объект Figure

ax1 = fig.add_subplot(2, 2, 1)  # Разбиваем график на 2x2. Выбираем первую ячейку
ax2 = fig.add_subplot('222')  # Разбиваем график на 2x2. Выбираем вторую ячейку 
ax3 = fig.add_subplot(223)  # Разбиваем график на 2x2. Выбираем третью ячейку 
````

**fig.add_subplot(2, 2, 1)** - это означает, что график будет поделен на 4 части, сеткой 2x2, и мы выберем первую часть. Части нумеруются сверху вниз, слева направо.

Аналогично работают и другие варианты. **matplotib** позволяет отправить аргументы так же строкой или одним числом. 

Способ передачи аргументов в виде строки (второй в коде предыдущего блока) в некоторых оболочках устарел. Старайтесь его избегать.

image.png

При добавлении каждого отдельного подграфика, график перекраивается заново.
Из-за этого если *figure* будет разбит неравномерно, то есть в разных **add_subplot** первые 2 цифры будут разные, то подграфики будут перекрываться друг с другом.

````python
fig = plt.figure()

ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 2, 1)
ax3 = fig.add_subplot(2, 2, 3)
````

В данном случае разбиение пытается пройти по следующему сценарию. Создается 4 квадратных подграфика и 2 прямоугольных подграфика. 
Скорее всего, это не тот результат, который намеревается получить программист.

С другой стороны, таким образом вы можете использовать `add_subplot` не только для деления на квадраты:

````python
fig = plt.figure()

ax1 = fig.add_subplot(1, 2, 2)  # Одна строка, два столбца. Выберем вторую ячейку.
ax2 = fig.add_subplot(2, 2, 1)  # Две строки, два столбца. Выберем первую ячейку.
ax3 = fig.add_subplot(2, 2, 3)  # Две строки, два столбца. Выберем третью ячейку.
````

## Стиль

**matplotlib** позволяет настраивать внешний вид графиков, изменять стили и прочие представления графов.

````python
plt.plot([1, 2, 3], color = 'red', marker = '^', linestyle = '--', linewidth = 2, markersize = 5)
````

- `color = 'red'`- цвет линии;
- `marker = '^'`  - тип маркера точек; `^` - означает треугольник, `o` - точку, `+`, `*` - плюс и звезду соответсвенно. При проектировании предполагалось, что значения параметра будут понятны интуитивно.
- `linestyle = '--'`- тип линии; `--` - прерывистая линия, `-` - непрерывная линия, `:` - пункитир.
- `linewidth = 2` - ширина линии;
- `markersize = 5` - размер маркера.

````python
# Альтернативный вариант
graf,  = plt.plot([1, 2, 3])  # graf, - распаковка массива из одной переменной
# plot возвращает массив с одним элементом
# запятая после graf нужна чтобы распаковать его
graf.set_color('red')
graf.set_marker('^')
graf.set_linestyle('--')
````

Запятая после переменной - это хитрый способ распаковать массив из одного элемента.

````python
# Пример обработки массива из одного элемента без распаковки и с распаковкой
a = [1]
print('Без запятой - ', a)
a, = [1]
print('С запятой - ', a)
````

Больше о стилях можно почитать тут: https://pyviy.blogspot.com/2009/09/matplotlib.html

Добавление легенды (описания) к графикам:

````python
plt.plot([1, 2, 3])
plt.plot([3, 2, 1])
plt.legend(['line1', 'line2'])  # Первому графику соответствует надпись "line1", второму "line2"
````

Если вы хотите добавить подпись к определенной линии, то можно передать методу `legend` два списка: в первом будут содержаться названия графиков, во втором подписи к ним.

````python
p1, = plt.plot([1, 2, 3])
p2, = plt.plot([3, 2, 1])
plt.legend([p1, p2], ['line1', 'line2'])  
````

Намного проще указать подпись к графику прямо при его создании с помощью аргумента **label**. Обратите внимание, что в этом случае все еще требуется вызов `plt.legend()` для отрисовки легенды.

````python
plt.plot([1, 2, 3], label='Line 1')  # Указываем легенду при инициализации графика с помощью параметра label
plt.plot([3, 2, 1], label='Line 2')
plt.legend()
````

## Типы графиков

По умолчанию **matplotlib** оставляет пустое место вокруг каждого подграфика и между подграфиками. Величина расстояния между подграфиками изменяется с помощью метода `subplots_adjust` объекта *Figure*:

```python
plt.subplots_adjust(left=None, bottom = None, right=None, top=None, wspace=None, hspace=None)
```
`left` - расстояние слева,
`right` - расстояние справа,
`bottom` - расстояние снизу, 
`top` - расстояние сверху,
`wspace` и `hspace` определяют, какой процент соответственно от ширины и высоты рисунка должен составлять промежуток между подграфиками. 

Рассмотрим пример:

````python
fig, axes = plt.subplots(2, 2, sharex = True, sharey=True)  # Создает 4 ячейки с общими осями

for i in range(2):
    for j in range(2):
        # Строим гистограммы нормально распределенного множества чисел. О функции hist() подробнее будет сказано ниже
        axes[i, j].hist([random.normalvariate(0, 10) for _  in range(500)], bins = 50, color = 'k', alpha=0.5)

plt.subplots_adjust(wspace = 0, hspace = 0)
````

Заметьте, что мы введи интересные способ разбиения на ячейки: `fig, axes = plt.subplots`

`sharex = True`, `sharey=True`, означает совместное использование осей x и y на графиках.

`plt.tight_layout()` метод для автоматического изменения размеров графиков так чтобы они они не мешали друг другу и помещались на изображение.

````python
# Точно такую же сетку мы уже создавали, но тогда не было применено выравнивание.
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 2)
ax2 = fig.add_subplot(2, 2, 1)
ax3 = fig.add_subplot(2, 2, 3)
plt.tight_layout()  #  Выполните код и сравните результат с тем, что выше
````

Для того, чтобы построить круговую диаграмму, можно воспользоваться функцией `pie()`.

`labels` задает метки;<br>
`sizes` задает размеры долей круга;<br>
`explode` задает расстояние, на которое отделяется кусок от всего круга;<br>
`autopct` задает формат, в котором выводится подпись доли куска;<br>
`shadow` задает отображение тени;<br>
`startangle` определяет угол поворота горизонтальной оси круга против часовой стрелки.

````python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # От круга будет отделен только второй кусок. 

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.2f%%', shadow=True, startangle=90)
plt.axis('equal')  # Равное соотношение сторон гарантирует, что круг будет отрисован именно в виде круга.
````

## Фигуры

В **matplotlib** так же можно отрисовывать произвольные геометрические фигуры при помощи [`patches()`](https://matplotlib.org/api/patches_api.html) и [`add_patch()`](https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.add_patch.html).

Функция `Rectangle()` строит прямоугольник, первый ее аргумент - *координаты левого нижнего угла*, второй - *ширина*, третий - *длина*, color - *цвет*, alpha - *прозрачность*. 

Функция `Circle()` строит окружность (или овал), первый ее аргумент - координаты центра, второй - радиус, color - цвет, alpha - прозрачность. 

Функция`Polygon()` строит многоугольник, первый ее аргумент - список вершин многоугольника, color - цвет, alpha - прозрачность. <br>

````python
fig, ax = plt.subplots(1)

# Строим окружность
rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color = 'k', alpha= 0.3)

# Строим окружность. На построенном графике эта окружнасть выглядит не как окружность, т.к. на осях координат разные длины единичных отрезков
circ = plt.Circle((0.7, 0.2), 0.15, color = 'b', alpha=0.3)

# Строим многоугольник
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], color= 'g', alpha=0.5)

# Добавляем фигуры на наш холст
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
````

## Конфигурация matplotlib

Параметры **matplotlib** можно посмотреть с помощью `rcParams`. Это словарь, так что можно легко посмотреть список всех элементов.

````python
plt.rcParams['figure.figsize']  # Размер холста в формате [длина,ширина]
````

````python
plt.plot([1,2,3])  # Изобразим простейшую линейную функцию y = x
````

Изменим размер холста:

````python
plt.rcParams['figure.figsize'] = (10, 10)  # Задаем новую длину и ширину холста
````

````python
plt.plot([1,2,3])  # Снова отрисуем линейную функцию y = x
````

Или так:

````python
plt.rc('figure', figsize= (10, 10))  # Альтернативный способ изменить размер холста
````

В программе параметры проще задать в виде словаря:

````python
# Зададим параметры для шрифта
font_options = {'family' : 'monospace', 
                'weight' : 'bold', 
                'size' : 12}
# Применим параметры
plt.rc('font', **font_options)  # ** - это способ передачи аргументов в функцию поэлементно
````

Вернем стандартные значения после всех предыдущих изменений:

````python
# Возвращаем старые размеры холста
plt.rcParams['figure.figsize'] = (6, 4)
# Возвращаем старые параметры шрифта
plt.rc('font', **{'family' : 'sans-serif',
                  'weight' : 'normal',
                  'size' : 10})
````

````python
import numpy as np  # Импортируем библиотеку numpy
````

Рассмотрим результат применения `np.meshgrid()`:

````python
x = [1, 2, 3]
y = [2, 5, 6, 8]
xgrid, ygrid = np.meshgrid(x, y)
print(f'xgrid=\n{xgrid},\nygrid=\n{ygrid}')
````

Теперь для любой пары индексов (i, j) легко можно получить координаты точки в плоскости xy:

````python
print((xgrid[1, 2], ygrid[1, 2]))  # = (3, 5)
print((xgrid[3, 0], ygrid[3, 0]))  # = (1, 8)
````

Хорошей практикой является разделения *этапа подготовки данных* от их визуализации. Создание **сетки** и расчет функции определим в отдельную функцию:

````python
def makeData ():
    # Строим сетку в интервале [-10, 10) с шагом 0.1 по обоим координатам
    x = np.arange (-10, 10, 0.1)
    y = np.arange (-10, 10, 0.1)

    # Создаем двумерную матрицу-сетку
    xgrid, ygrid = np.meshgrid(x, y)
    
    # В узлах рассчитываем значение функции? np.sin - синус
    zgrid = np.sin (xgrid) * np.sin (ygrid) / (xgrid * ygrid)
    return xgrid, ygrid, zgrid
````

````python
# Получаем значения для x, y, z
x, y, z = makeData()
````

````python
# Инициализируем холст
fig = plt.figure()
# Добавляем подграфик в график
ax = fig.add_subplot(111, projection='3d')
# Отрисовываем график на холсте
ax.plot_surface(x,y,z)
````

`plot_surface(self, X, Y, Z, *args, **kwargs)` - задает поверхность из точек с координатами записанными в массивах X Y Z.

| Аргументы | Тип | Описание |
| --- | --- | --- |
| `X, Y, Z` | `2d arrays` | Массивы с данными об X, Y, Z координатах соответственно. |
| `rcount, ccount` | `int` | Максимальное количество отсчетов в каждом направлении. Если входные данные больше, то все что дальше отрезается. По умолчанию 50.|
| `rstride, cstride` | `int` | Задают шаг вывода графика. Чем меньше шаг, тем точнее отображается функция, но тем дольше происходит рисование. |
| `color` | `color-like` | Цвет графика.|
| `cmap` | `Colormap` | Градиент цветов, чтобы цвет ячейки графика зависел от значения функции в этой области.|
| `norm` | `Normalize` | Нормализация цветовой палитры. |
| `vmin, vmax` | `float` | Границы нормализации. |
| `**kwargs` | \_  |  Other arguments.|

````python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# Задаем шаг вывода графика и цвет
ax.plot_surface(x, y, z, rstride=10, cstride=10, color='g')
````

````python
from matplotlib.colors import LinearSegmentedColormap  # Класс matplotlib, содержащий цвета
````

Параметры метода **from_list():**

- Имя создаваемой карты 
- Список цветов, начиная с цвета для минимального значения на графике (голубой - `b`), через промежуточные цвета (у нас это белый - `w`) к цвету для максимального значения функции (красный - `r`). 
- Количество цветовых переходов. Чем это число больше, тем более плавный градиент, но тем больше памяти он занимает. 

````python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# Зададим поверхность с новым изученным методом from_list
ax.plot_surface(x, y, z, rstride=3, cstride=3, 
                cmap = LinearSegmentedColormap.from_list("red_blue", ['b', 'w', 'r'], 256))
````

**Результат можно повернуть, задав значения elev, azim в ax.view_init(elev, azim):** https://proproprogs.ru/modules/matplotlib-kak-stroit-trehmernye-grafiki
- *elev* хранит угол места в плоскости z (в градусах).
- *azim* хранит угол азимута в плоскости (x, y) (в градусах).

Пример:

````python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, rstride=3, cstride=3, 
                cmap = LinearSegmentedColormap.from_list("red_blue", ['b', 'w', 'r'], 256))
ax.view_init(10, 70)
````

Отрисовка линии в 3х мерном пространстве:

````python
fig = plt.figure()
ax = fig.gca(projection='3d')
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)
ax.plot(x, y, z, label='parametric curve')
ax.legend()
````

Отрисовка точек в трехмерном пространстве:

````python
def randrange(n, vmin, vmax):
    '''
    Вспомогательная функция для создания массива случайных чисел формы (n, )
    с каждым распределенным числом Uniform(vmin, vmax).
    '''
    return (vmax - vmin)*np.random.rand(n) + vmin

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

n = 100

# Для каждого набора настроек стиля и диапазона отображает n случайных точек в поле
# x на отрезке [23, 32], y на отрезке [0, 100], z на отрезке [zlow, zhigh].
for c, m, zlow, zhigh in [('r', 'o', -50, -25), ('b', '^', -30, -5)]:
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zlow, zhigh)
    ax.scatter(xs, ys, zs, c=c, marker=m)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
````
