Строки кода, которые относятся к одному и тому же элементу программы, должны быть на одном уровне отступа.

1. В качестве отступа можно использовать 2 или 4 пробелла или *Tab* (не рекомендуется). При этом следует придерживаться одного формата во всей программе.
2. Длинные инструкции можно распологать на нескольких строках используя символ продолжения строки:

```python
import math
a = math.sqrt(9) + \
    math.pow(9, 2)
a
```

3. Если была открыта скорбка (, {, [ или тройные кавычки """, то строка будет продолжаться, пока не появится соответсвующий закрывающий символ.

```python
a = [1,
    2,

    3]

b = """a

b
c"""
print(a, b, sep='\n')  # sep - это разделитель. В данном случае печать каждого аргумента будет отделена переносом строки
```

Алгоритм выполнения программы, как правило, определяется выполнением различных условий. Если условие выполняется, то выполняется один блок кода, а если не выполняется, то другой.
Для реализации этого механизма в **python** есть условный оператор.

## Сравнение объектов. Условный оператор if
| Оператор | Значение |
| --- | --- |
| < | строго меньшне чем |
| <= | меньше или равно |
| > | строго больше |
| >= | больше или равно |
| == | равно |
| != | не равно |
| is | сравнение объектов |
| is not | отрицание сравнения объектов |

Запись условного оператора начинается с ключевого слова **if**, псоле которого указывается условие. Результат проверки условия возвращает значение типа **boolean**.
**Boolean** - это тип данных, который может принимать 2 значения: **True (1)** или **False (0)**.
Если результат проверки **True**, то выполняется код, который находится под блоком **if**.

Оператор **not** опровергает логическое после него. `not True` -> `False`, `not False` -> `True`.

На случай, если результат проверки условия if вернет False, то можно либо завершить работу условоного оператора, либо написать
другое условие с помощью ключевого слова elif, либо написать блок else.

Пример оператора if:

```python
a = 5  # Инициализируется переменная
if a > 0:
    print('a - положительное')  # Эта строка выполнится, если проверка условия вернет результат True
elif a == 0:  # Эта строка выполнится, если проверка первого условия вернет результат False
    print('a - равно 0')
else:  # Эта строка выполнится только в случае, если проверки первых двух условий вернут результат False
    print('a - отрицательное')
```

Обратите внимание, что для проверки условия используется оператор **==** (двойное равно), а для присвоения используется **=** (одинарное равно).
С точки зрения **Python** - это два разных оператора. Одна из частых ошибок - написание оператора сравнения с одинарным равно.
Также в условном операторе используются отступы (например, 4 пробела), которые позволяют отделить набор строк кода, которые выполняются при выполнении условия, о них будет сказано ниже.

В **python** переменные принимают **True** и **False** значения в условных операторах, так что можно, и рекомендуется, не использовать явно операторы сравнения.
### Плохой код:

```python
a = True  # Присваиваем переменной a значение True. Тип данных Boolean.
if a == True:  # "Плоходй код", т.к. такая строчка приведет к сравнению True == True
    print('a is True')
```

### Хороший код:

```python
a = True
if a:  #  "Хороший код"
    print('a is True')
if not a:
    print('a is not True')
```

Таким способом можно проверять переменные, имеющие **True** или **False** значения:

```python
a = ''  # a - пустая строка
if a:
    print(a)
if not a:  # Здесь результат проверки будет обратным результату проверки a. Если "a" - True, то "not a" вернет False и наоборот.
    print('a is empty')

b = 0
if b:
    print('b is True')
if not b:
    print('b is False')

c = 1
if c:
    print('c is True')
if not c:
    print('c is False')
```

**is** и **==** - это не одно и то же. Оператор **==** проверяет на равенство значений объектов, а **is** проверяет на совпадение объектов (пояснение в блоке немного ниже).

Объект, имеющий "непустое значение", т.е. неравный None, 0, '' (пустая строка), соответсвует логическому значению Истина (**True**).

Если необходимо именно значение **True**:

```python
a = 1
print('a =', a)
if a == True:
    print('a == True')
else:
    print('a != True')

# В данном случае is проверяет не значение "a", а ее ссылку на объект True.
if a is True:
    print('a is True')
else:
    print('a is not True')
```

Тип данных None означает отсуствие значения. Это не то же самое, что 0 или пустая строка.
0 или пустая строка могут нести какой-то смысл (например, температура 0 градусов).
В то же время None означает отсутствие значения как такового.
Проверка на значение None:

```python
a = None
if a is None:
    print('a is None')
```

Условные операторы можно писать в одну строку в тех случаях, когда подразумеваются короткие конструкции, которые
можно описать лаконично.
Сначала пишется то, что выполнится в случае **True**, затем ключевое слово **if** и условие, затем ключевое слово **else** и код, который выполнится в случае **False**.
Пример *однострочного* **if**:

```python
a = 1 if True else 2
print(a)
```

## Циклы: всего 2 типа
Цикл - это блок кода, содержимое которого (тело цикла) может выполняться больше одного раза.
#### Цикл **while**
Цикл **while** выполняется до тех пор, пока выполняется заданое условие.
В начале каждой итерации проверяется выполнение условия. Если проверка возвращает **истину** (**True**), то выполняется тело цикла.
Если проверка возвращает **ложь** (**False**), то выполняется блок кода **else**.
Теоретически цикл **while** может быть бесконечным, однако крайне не рекомендуем писать подобный код (за исключением отдельных случаев, например, построения сокета).
Цикл **for** выполняется для заданного количества элементов из набора (списка, множества, ...).

Ключевое слово **continue** прерывает текущую итерацию цикла и переходит на следующую.
Ключевое слово **break** прерывает выполнение кода в цикле и выводит из него. Оператор **else** в таком случае не выполняется, т.е. **else** работает так же, как и в условном операторе.
Проверьте, как и когда выполняются условия **continue**, **break**, **else**:

```python
a = 0
while a < 5:
    a += 1
    print(a)
    if a == 2:  # Пропустить ход если а == 2
        continue
    if a == 4:  # Выйти из цикла если a == 4
        break
    print('continue ', a)
else:
    print('all out')  # Строка не выполнится, так как при a=4, выполнится break и else будет пропущен
```

В случае, когда функции **range** передаются два аргумента, аргументы означают начало диапазона чисел (включительно) и конец диапазона (не включительно)

В случае, когда функции **range** передаются три аргумента, аргументы означают начало диапазона чисел (включительно), конец диапазона (не включительно) и шаг увеличения (уменьшения) чисел в последовательности.


```python
# Данный цикл отработает в диапазоне от 5 до 15 с шагом 3
for i in range(5, 15, 3):
    print("Вывожу число: " + str(i))
print("Конец третьего цикла")
```

## Пример: отфильтровать все элементы со значением больше 2.

Способ 1. Стандартный цикл:

```python
a = [1, 2, 3, 4]
b = []
for i in a:
    if i > 2:
        b.append(i) # Метод append() добавляет новый элемент в список
print(b)
```

Способ 2. Списковое включение.

В примере ниже создается список из значений списка a, которые больше 2:

```python
a = [1, 2, 3, 4]
b = [i for i in a if i > 2]  # Циклы также будут рассмотрены позднее
print(b)
```

Функция `filter()` подробнее рассматривается в следующих блокнотах, к способу 3 можно вернуться после этого или изучить эту тему самостоятельно.

Способ 3. Использование функции `filter()`:

```python
a = [1, 2, 3, 4]
b = list(filter(lambda x: x > 2, a))  # lambda функции рассматриваются позднее
print(b)
```

## Пример. Увеличить значение каждого элемента на 3.

Способ 3. Использование функции `map()`:

```python
a = [1, 2, 3]
a = list(map(lambda x: x + 3, a))
print(a)
```

###### get(key [, defval])
Если ключ `key` присутствует в словаре, то метод возвращает значение, соответствующее этому ключу. Если ключ отсутствует, то возвращается значение `None` или значение из второго параметра `defval` (дефолтное значение)

```python
d = {'a': 1, 'b': 2}
print(d.get('a'))
print(d.get('c'))
```

###### setdefault(key [,defval])
Если ключ присутствует в словаре, то метод возвращает значение, соответствующее этому ключу. Если ключ отсутствует, то вставляется новый элемент со значением, указанным во втором параметре. Если второй параметр не указан, то `None`:

```python
d = {'a': 1, 'b': 2}
print(d.setdefault('a'))

d.setdefault('d', 3)
print(d)
```

```python
d = {'a': 1, 'b': 2}
d.pop('a')
```

## underscore

**`_` (подчеркивание, underscore)** - специальная переменная в **python**. Она используется для сохранения результата последней выполненой команды и, в том числе, для обозначения переменных, которые не будут использоваться:

```python
for _ in range(5):
    print("I can use underscore")
```

Забегая сильно вперед, немного коснемся темы объектно-ориентированного программирования. При именовании переменных двойной underscore обычно используют для обозначения функций и переменных, которые будут использоватся только внутри класса.

```python
class A():  # Инициализируем класс A
    def __init__(self):  # Инициализируем конструктор класса, который вызывается при инициализации экземпляра класса
        self._var = 'su'
        self.__var = 'du'

a = A()  # Инициализируется экземпляр класса А
print('a._var = ', a._var)
```

```python
try:  # Это механизм исключений. Он будет объяснен позже
    a.__var  # AttributeError: 'A' object has no attribute '__var'
except AttributeError:
    print('no such attribute as __var')

print(a._A__var)
```

## Исключения

Некоторые участки кода могут выполниться с ошибкой. Если никак не обработать появление ошибки, то выполнение программы прекратится. Однако с помощью операторов **try** - **except** - **else** - **finally** можно повернуть работу программы в другое русло, и ее выполнение продолжится.
Совокупность данных операторов называется механизмом исключений.
Исключение, возникающее при работе программы, зависит от типа ошибки.

*Агоритм выполнения*.  Выполняется код, находящийся в теле под ключевым словом **try**. Возможно два варианта:
1.   Этот код не вызвал ошибку, тогда выполняется код в теле под ключевым словом **else**, затем выполняется код в теле под ключевым словом **finally**.
2.   Этот код вызвал ошибку, тогда выполняется проверка на соотвествие типа этой ошибки тому типу, который указан после ключевого слова **except**. Если такое совпадение найдено, то выполянется код в теле под ключевым словом **except**. Затем выполняется код в теле под ключевым словом **finally**.

Список всех встроеных исключений: https://docs.python.org/3/library/exceptions.html


В качестве упражнения, попробуйте поизменять значения ключа *value* в коде:

```python
my_dict = {"a": 1, "b": 2, "c": 3}

try:
    value = my_dict["z"]
except IndexError:  # Исключение, возникающее в случае обращения к несуществующему индексу
    print("This index does not exist!")
except KeyError:  # Обращение к несуществующему ключу словаря
    print("A KeyError occurred!")
else:
    print("No error occurred!")
finally:
    print("The finally statement ran!")
```

```python
try:
    a = 1 / 0
except ZeroDivisionError as e:  # Деление на ноль - это тоже исключение
    print('Get error:')
    print(e)
```

## Функции
*Функция* - фрагмент программного кода, к которому можно обратиться из другого места программы.
Суть функции в том, чтобы получать на вход какие-то аргументы (возможно пустое множество аргументов),
выполнить какие-то действия и передать с помощью **return** результат своей работы обратно программе.


```python
def func(positional, keyword=value, *args, **kwargs):
    pass
```

+ **positional** - Позиционные аргументы. Необязательны, не имеют значений по умолчанию
+ **keyword** - Аргументы с ключевым словом. Необязательны, имеют значение по умолчанию
+ **args** - Список с произвольным количеством аргументов. Необязателен, не имеет значений по умолчанию
+ **kwargs** - Словарь с произвольным количеством аргументов. Необязателен, не имеет значений по умолчанию

Оператор `return` возвращает в то место, откуда вызывается функция, значение, которое идет следом за `return`:

Переменные, объявленные в теле функции, называются *локальными* и доступны только в этой функции. Переменные объявленные в теле основной программы являются *глобальными* при выполнении функции.

Без предварительного объявления *глобальных* переменных в функции можно использовать их значения, однако эти значения нельзя изменить.
Для изменения значения *глобальных* переменных их необходимо объявить в функции,используя ключевое слово `global`.

## Декоратор @


обертка для других функций или класса.
Основная задача декораторов - избежать повторения одного и того же кода. Рассмотрим пример:

```python
# Объявление декоратора:
def makebold(fn):
    def wrapped():
        return '<b>' + fn() + '</b>'
    return wrapped

# Объявление декоратора:
def makeitalic(fn):
    def wrapped():
        return "<i>" + fn() + "</i>"
    return wrapped

# Применение двух декораторов к результатам функции hello():
@makebold
@makeitalic
def hello():
    return 'hello world'

hello()
```

## type, isinstance

С помощью `type()` можно получить тип объекта или проверить, принадлежит ли объект какому-то типу:

```python
a = 10
print(type(a))
print(type(a) == int)
```

То же самое можно сделать с помощью функции isinstance().

```python
a = 10
isinstance(a,int)
```

Но isinstance так же поддерживает наследование и может проверить принадлежность хоть одному классу из списка.
Подробнее о наследовании будет сказано в следующих блокнотах.

```python
class A(list):
    pass
a = A()
print(type(a) == list)
print(isinstance(a, list))
```

## assert

Проверяет выполнение условия. Если не выполняется, то возбуждается AssertionError.

```python
assert False
```

Класс **TestCase** из модуля **unittest** представляет некоторое количество методов для тестирования и сообщения об ошибках:

| Метод | Проверяет что | Начиная с |
| --- | --- | --- |
| assertEqual(a, b) | a == b | \_ |
| assertNotEqual(a, b) | a != b | \_ |
| assertTrue(x) | bool(x) is True | \_ |
| assertFalse(x) | bool(x) is False | \_ |
| assertIs(a, b) | a is b | 2.7 |
| assertIsNot(a, b) | a is not b | 2.7 |
| assertIsNone(x) | x is None | 2.7 |
| assertIsNotNone(x) | x is not None | 2.7 |
| assertIn(a, b) | a in b | 2.7 |
| assertNotIn(a, b) | a not in b | 2.7 |
| assertIsInstance(a, b) | isinstance(a, b) | 2.7 |
| assertNotIsInstance(a, b) | not isinstance(a, b) | 2.7 |

```python
import unittest
test = unittest.TestCase()
a = .01
b = .11 - .1
print('a =', a)
print('b =', b)

test.assertAlmostEqual(a, b)  # Проверка на _почти_ равенство, проходит
print('Проверка на полное равенство, не проходит (т.к. возникает погрешность вычисления. Для решения этой проблемы можно использовать модуль decimal')
test.assertEqual(a, b)
```
