## Сравнение вещественных чисел
Теперь давайте рассмотрим практические примеры.

Зададим две переменные, с виду равные по значению:

```python
a = 0.1 + 0.1 + 0.1
b = 0.3
```

Как вы думаете, равны ли a и b?

```python
print("Равны" if a == b else "Не равны")
```

Почему так происходит? Выведем a и b:

```python
print(f'a = {a}')  
print(f'b = {b}')
```

a и b не равны как раз из-за особенностей *представления* **вещественных чисел** в памяти компьютера.
Сравнение целых чисел с вещественными тоже не всегда работает корректно.

Зададим две с виду равные переменные: одна вещественная (тип float), другая целая (тип int): 

```python
a = 25.0**100 
b = 25**100  
```

Будут ли равны переменные?

```python
print("Равны" if a == b else "Не равны")
```

Почему так происходит? Выведем a и b:

```python
print(f'a = {a}')  
print(f'b = {b}')
```

Однако в *отдельных случаях* проверять целые числа на равенство с вещественными можно. 

```python
a = 5
b = 5.0
print("Равны" if a == b else "Не равны")
```

Способ выше работает только, если число было задано и с ним не производилось никаких операций. В программировании мы очень часто не знаем, были ли произведены операции с числом или нет.<br>
Как же все-таки сравнивать вещественные числа? Можно сравнивать числа *приблизительно*. Для этого зададим какое-нибудь маленькое число **delta** и будем считать, что если разницу между двумя вещественными числами меньше чем **delta**, то они равны.

```python
delta = 0.000000001  # Допустимая разница между вещественными числами
a = 0.1 + 0.1 + 0.1
b = 0.3
print(abs(a-b) < delta)  # True. Числа равны!
```

Функция **abs** находит модуль числа. Так как мы не знаем какое из чисел больше, то советуем брать модуль от разницы. Если этого не сделать, то возможны подобные **казусы**:


```python
delta = 0.000000001  # Допустимая разница между вещественными числами
a = 0.1 + 0.1 + 0.1
b = 1000
print(a-b < delta)  # True. Числа равны! Как же так?
```

Повторим тот же код, но уже применяя **abs**.

```python
delta = 0.000000001  # Допустимая разница между вещественными числами
a = 0.1 + 0.1 + 0.1
b = 1000
print(abs(a-b) < delta)  # False. Числа не равны.
```

## Представление вещественных чисел
Метод **as_integer_ratio** возвращает пару целых чисел, отношение которых точно равно исходному вещественному числу с положительным знаменателем. Первое число - числитель, второе - знаменатель.<br>
Другими словами **as_integer_ratio** позволяет представлять число в виде *обыкновенной дроби*.

```python
a = 0.1 + 0.1 + 0.1
b = 0.3
# Представление одинаковое, но числа не равны
print(a.as_integer_ratio())  # (1351079888211149, 4503599627370496)
print(a.as_integer_ratio())  # (1351079888211149, 4503599627370496)
```

## Округление вещественных чисел
В различных ситуациях могут потребоваться различные правила, по которым должно округляться число, рассмотрим основные из них. <br>
**int()** - округление путем отбрасывания вещественной части;<br>
**floor()** - округление в сторону уменьшения;<br>
**round** - математическое округление. Если число заканчивается на .5, то округляется в сторону ближайшего четного числа.<br>
Изучите примеры:

```python
import math  # Предоставляет некоторые функции округления
print(f' int(2.4) = {int(2.4)}') 
print(f' int(2.6) = {int(2.6)}') 
print(f' int(2.5) = {int(2.5)}') 
print(f' int(3.5) = {int(3.5)}') 
print(f' int(-2.5) = {int(-2.5)}') 
```

```python
print(f'math.floor(2.5) = {math.floor(2.5)}') 
print(f'math.floor(3.5) = {math.floor(3.5)}') 
print(f'math.floor(-2.5) = {math.floor(-2.5)}')  
```

```python
print(f'round(2.5) = {round(2.5)}')
print(f'round(3.5) = {round(3.5)}')
print(f'round(-2.5) = {round(-2.5)}')
```

```python
print(f'round(2.9) = {round(2.9)}')
print(f'round(3.1) = {round(3.1)}')
print(f'round(-2.55) = {round(-2.55)}')
```
